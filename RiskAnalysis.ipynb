```
import numpy as np
import pandas as pd
```

```
train_data=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/train7.csv')
test_data=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/test7.csv')
```

```
train_data.head()
```

```
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
train_data=scaler.fit_transform(train_data)
```

```
x_data=train_data[:,:23]
y_data=train_data[:,23:]
  
```

```
x_data
```

```
y_data.shape
```

```
import keras
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split
import tensorflow as tf
from keras.optimizers import SGD  
```

```
x_train,x_test,y_train,y_test = train_test_split(x_data,y_data,test_size = 0.2)
```

```
x_test.shape
```

```
y_data[0:5]
```

```
model=keras.Sequential([
    keras.layers.Dense(64,activation='relu',input_dim=23),
    keras.layers.Dense(32,activation='relu'),
    keras.layers.Dense(32,activation='relu'),
    keras.layers.Dense(16,activation='relu'),
    keras.layers.Dense(8,activation='relu'),
    keras.layers.Dense(4,activation='relu'),
    keras.layers.Dense(1,activation='sigmoid')
])
```

```
sgd=SGD(lr=0.0099,decay=1e-6)
model.compile(optimizer=sgd,
              loss='binary_crossentropy',
              metrics=['accuracy'])
```

```
model.fit(x_data, y_data, epochs=10, batch_size=32)
```

```
test_data.shape
```

```
[loss, acc] = model.evaluate(x_test,y_test)
```

```
test_data=scaler.fit_transform(test_data)
```

```
y_pred=model.predict(test_data)
```
